```python
# Displaying rows of the dataset using head() function
# By default head() will display 5 rows
# T denotes the transpose of the table (rows becomes columns and columns becomes rows)
airbnb_df.head().T
```

The number of rows before dropping duplicates and number of rows after dropping duplicates are same. Hence, there is not any duplicates in the dataset.

The columns **name** & **host_name** have very few null values which can handled easily by replacing null values with some values.

***Handling missing data***

```python
# Filling missing values in name & host_name with "No name"
airbnb_df["name"].fillna ('no name', inplace = True)
airbnb_df["host_name"].fillna ("no name", inplace = True)
airbnb_df.isnull().sum()
```

```python
# Concating the columns (number_of_reviews, reviews_per_month and last_review) where number_of_reviews = 0 & reviews_per_month = 0
new_num_of_reviews_df = airbnb_df[airbnb_df ["number_of_reviews"] == 0]
new_reviews_df = airbnb_df[airbnb_df ["reviews_per_month"]==0]
new_df = pd.concat([new_num_of_reviews_df,new_reviews_df], axis = 0)
new_df [["number_of_reviews","reviews_per_month","last_review"]]
```

From the above table, we can see that there are 10052 rows where the **reviews_per_month** column has null values and **number_of_reviews** = 0. So, we can say that **reviews_per_month** = 0 when **number_of_reviews** = 0. The missing values of **reviews_per_month** can replaced by 0 as there is no reviews for such listing.

```python
# Replacing all the null values in review_per_month column by 0
airbnb_df["reviews_per_month"] = airbnb_df["reviews_per_month"].replace(np.nan, value=0)
airbnb_df.isnull().sum()
```

The **last_review** column can be dropped as it tell us about the last date of the review, but when **number_of_reviews** = 0 the value of **last_review** will always be 0 and it will not be required in the analysis

```python
#Dropping last_review column using drop() function
airbnb_df = airbnb_df.drop(['last_review'], axis=1)
airbnb_df.head().T
```

8. **The price distribution between different neighbourhood groups**

**Dealing with outliers**

```python
#Using describe() function for importing statistical details
airbnb_df [["price"]].describe().T.round(2)
```

```python
#Calculating Inter Quartile Range (IQR)
#IQR = 75% - 25%
percentile_75 = airbnb_df['price'].quantile(0.75)

percentile_25 = airbnb_df['price'].quantile(0.25)

iqr = percentile_75 - percentile_25

print ("IQR = ", iqr)
```

```python
# Upper limit & lower limit of the data points
upper_limit = percentile_75 + 1.5*iqr
lower_limit = percentile_25 - 1.5*iqr
print ("Upper limit: ", upper_limit,"\n", "Lower Limit: ", lower_limit)
```

```python
# Price should be less than or equal to upper limit
cond_1 = (airbnb_df['price'] <= upper_limit)
# Price should be greater than lower limit
cond_2 = (airbnb_df['price'] > lower_limit)
airbnb_df = airbnb_df[ cond_1 & cond_2]
```

```python
# Visualization of the data
ax = sns.boxplot (x = "neighbourhood_group", y = "price", data = airbnb_df)
ax.set_title ("Price distribution between different neighbourhood groups")
plt.show()
```

**Insights:**

* Box plot is used when the comparison in distributions needs to be performed between groups. The box plot represents the minimum, maximum, median, first quartile, and third quartile in the data set. It is also useful in comparing the distribution of data across data sets by drawing boxplots for each of them.

* The median of the price in **Manhattan** group is maximum as compared to other groups. The median value is denoted by the center line of each boxplot.

* The **Manhattan** group also has the maximum price denoted by the top line of the boxplot whereas **Staten Island** has the minimum price denoted by the bottom line of the boxplot.
